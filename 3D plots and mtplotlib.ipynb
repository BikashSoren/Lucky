```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(20,12))


plt.figure(figsize=(10,8))
ax1.bar(d1['Name'],d1['SQL'],color='lightgreen',edgecolor='black')
ax1.set_title("Name vs SQL",fontsize=15,color='blue')
ax1.set_xlabel("Name")
ax1.set_ylabel("SQL")
ax1.grid()



ax2.plot(d1['Name'],d1['Python'],color='cyan',marker='^',markersize=10,label='Python')
ax2.plot(d1['Name'],d1['Statistics'],color='Red',marker='o',markersize=10,label='Statistics')
ax2.plot(d1['Name'],d1['SQL'],color='purple',marker='*',markersize=10,label='SQL')
ax2.set_title("Name vs Python vs Statistics vs SQL",fontsize=15,color='blue')
ax2.set_xlabel("Name")
ax2.set_ylabel("Marks")
ax2.grid()
ax2.legend()

plt.figure(figsize=(5,3))
ax3.pie(x=d2['Total_marks'],labels=d2['Name'],autopct='%.2f%%',
       colors=['red','blue','yellow','green','cyan','deeppink','dodgerblue','orange'],
       explode=(0,0,0,0,0,0.2,0,0.1),shadow=True,counterclock=False)
ax3.set_title('Names',fontsize=15,color='blue')
ax3.legend(loc='right',bbox_to_anchor=(0.7,1,-0.8,-0.5))

ax4.stackplot(d4['Name'],d4['SQL'],d4['Python'],d4['Statistics'],labels=['Python','SQL','Statistics'])
ax4.set_xlabel('Name')
ax4.set_ylabel('Marks')
ax4.set_ylim(0,450)
ax4.legend()
ax4.set_title('Name vs Marks',fontsize=15,color='blue')

plt.show()
```

```python
fig=plt.figure(figsize=(5,6))
ax=fig.add_subplot(projection='3d')
ax.scatter(d4['Python'],d4['SQL'],d4['Statistics'])
ax.set_xlabel('Python')
ax.set_ylabel('SQL')
ax.set_zlabel('Statistics')
plt.show()
```
